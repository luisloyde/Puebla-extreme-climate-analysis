```python
from meteostat import Stations, Daily
from datetime import datetime
import numpy as np
```

# Obtaining data

The first step is to choose the station in the state of Puebla, Mexico.

```python
stations = Stations()
stations = stations.region('MX','PUE')
stations = stations.fetch(stations.count())

print(stations)
```

We conclude that Puebla, Pue. Station with WMO code 76685 is the station located at Puebla City, the study site. Now we analyze the coverage of the data to choose the appropiate period.

```python
start = datetime(1974, 1, 1)
end = datetime(2024, 1, 1)

data = Daily('76685', start=start, end=end)
coverage_50_years = data.coverage()

start = datetime(1994, 1, 1)
end = datetime(2024, 1, 1)

data = Daily('76685', start = start, end = end)
coverage_30_years = data.coverage()

print('For 50 years the coverage is ' + '{:.2f}'.format(coverage_50_years * 100) 
      + '% and for 30 years is ' + '{:.2f}'.format(coverage_30_years * 100) + '%')
```

Since 30-years period has a suitable coverage in comparison to 50-years period, we normalize data. This step is esential for inteporlation and filling gaps.

```python
start = datetime(1994, 1, 1)
end = datetime(2024, 1, 1)

data = Daily('76685', start=start, end=end)
data = data.normalize()
data = data.fetch()

data.head()
```

Now that the normalization is completed, we proceed to interpolate the data.

```python
data = data.interpolate()
data.head()
```

```python
np.mean(data.prcp)
```

# Data processing

## Precipitation

```python
precipitation_threshold = np.percentile(data['prcp'].dropna(), 99.9452)
print(precipitation_threshold)
extreme_precipitation = data[data['prcp'] > precipitation_threshold]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(extreme_precipitation.index, extreme_precipitation['prcp'], color='blue', label='Extreme precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.title('Puebla Extrem Precipitation Events (1980-2024)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.index, data['prcp'], color='gray', label='Mean precipitation')
plt.scatter(extreme_precipitation.index, extreme_precipitation['prcp'], color='red', label='Extreme precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.title('Precipitation over time')
plt.legend()
plt.show()
```

## Temperature

```python
temp_threshold = np.percentile(data['tmax'].dropna(), 99.9452)
print(temp_threshold)
extreme_temp_max = data[data['tmax'] > temp_threshold]
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.index, data['tmax'], color='gray', label='Maximum mean temperature')
plt.scatter(extreme_temp_max.index, extreme_temp_max['tmax'], color='red', label='Maximum extreme temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Temperature over time')
plt.legend()
plt.show()
```

```python
temp_threshold = np.percentile(data['tmin'].dropna(), 100-99.9452)
print(temp_threshold)
extreme_temp_max = data[data['tmin'] < temp_threshold]

plt.figure(figsize=(10,6))
plt.plot(data.index, data['tmin'], color='gray', label='Minimal mean temperature')
plt.scatter(extreme_temp_max.index, extreme_temp_max['tmin'], color='red', label='Minimal extreme temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Temperature over time')
plt.legend()
plt.show()
```

```python
temp_threshold = np.percentile(data['tavg'].dropna(), 99.9452)
print(temp_threshold)
extreme_temp_max = data[data['tavg'] > temp_threshold]

plt.figure(figsize=(10,6))
plt.plot(data.index, data['tavg'], color='gray', label='Mean temperature')
plt.scatter(extreme_temp_max.index, extreme_temp_max['tavg'], color='red', label='Maximum extreme temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Temperature over time')

temp_threshold = np.percentile(data['tavg'].dropna(), 100-99.7260274)
print(temp_threshold)
extreme_temp_min = data[data['tavg'] < temp_threshold]

plt.scatter(extreme_temp_min.index, extreme_temp_min['tavg'], color='blue', label='Minimal extreme temperature')
plt.legend()

plt.show()
```
